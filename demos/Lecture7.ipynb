```julia
using Plots, ComplexPhasePortrait, ApproxFun, SpecialFunctions
```

$$
\theta_j = 2\pi j/N
$$

```julia
N = 1000
θ = 2π/N * (0:N-1)
f = θ -> airyai(θ-5)
g = range(0,2π; length=1000)
plot(g, f.(g))
plot!([θ; 2π], [f.(θ); f(0)])
```

Non-periodic functions, trapezium rule converges slow:

```julia
f = θ -> airyai(θ-5)
Q_ex = sum(Fun(f,0..2π))
Q_T = 2π/N * sum(f.(θ))
Q_T - Q_ex
```

Periodic functions, trapezium rule converges fast:

```julia
f = θ -> 1/(2+cos(θ))
N = 50
θ = 2π/N * (0:N-1)
plot(g, f.(g))
plot!([θ; 2π], [f.(θ); f(0)])
```

```julia
Q_ex = sum(Fun(f,0..2π))
```

```julia
errs = [(θ = 2π/N * (0:N-1); Q_T = 2π/N * sum(f.(θ)); abs(Q_T - Q_ex)) for N=1:29]
plot(1:length(errs),
    errs; yscale=:log10)
```
