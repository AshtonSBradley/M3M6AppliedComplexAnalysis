```julia
using Plots, ComplexPhasePortrait, SpecialFunctions
```

```julia
f = z -> zeta(z)
u = z -> real(f(z))
v = z -> imag(f(z))

xx = range(-2,2; length=200)
yy = range(-2,20; length=200)

plot(contourf(xx, yy, u.(xx' .+ im*yy)),
        contourf(xx, yy, v.(xx' .+ im*yy)))
```

```julia
f = z -> sin(z)

r = z -> abs(f(z))
θ = z -> angle(f(z))

xx = range(-4,4; length=200)
yy = range(-2,2; length=200)

plot(contourf(xx, yy, r.(xx' .+ im*yy)),
        contourf(xx, yy, θ.(xx' .+ im*yy)))
```
